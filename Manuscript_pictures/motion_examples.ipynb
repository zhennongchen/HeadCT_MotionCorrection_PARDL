# motion examples

```python
gt = np.array([0, 0, -3, 0,0])

t = np.linspace(0, 100, 5, endpoint=True) # gantry rotation time

x = np.asarray(gt).reshape(-1)

spline_x = transform.interp_func(t,x)

N = 25
tmin, tmax = t.min(), t.max()
tt = np.linspace(tmin, tmax, N)

fig = plt.figure(figsize = (7,3))
plt.scatter(t, x, c = 'blue', marker = 'o', s = 50)
plt.plot(tt, spline_x(tt).reshape(-1), 'r-', label='B-spline fit', linewidth = 2.5)


plt.yticks(np.arange(0,-5.1,step = -1), fontsize = 16)
plt.xticks(np.arange(0,101,step = 20), fontsize = 16)
# plt.ylabel("translation (mm)", fontsize = 13)
# plt.xlabel("% of gantry rotation time", fontsize = 12)
plt.grid(True, linestyle='dashed', alpha=0.5)

save_path = '/mnt/mount_zc_NAS/motion_correction/pictures/Manuscript'
plt.savefig(os.path.join(save_path, 'B-spline-motion-example2.eps'), transparent = True)
```
